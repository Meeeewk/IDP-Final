```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd

def main():
    result = []
    url_start = 'https://nflcombineresults.com/nflcombinedata.php?year='
    url_end = '/&pos=&college=&proday=off'
    
    for year in range(2000, 2024):
        result = result + get_data_year(url_start + str(year) + url_end)
        time.sleep(random.randint(1, 3))
    df = pd.DataFrame(result)
    df.to_csv('combine_data')

def get_data_year(link):
    page = requests.get(link, verify=False)
    soup = BeautifulSoup(page.content, 'html.parser')
    table_rows = soup.select('#datatable > table > tbody tr')
    types_data = ['year', 'name', 'college', 'pos', 'height', 'weight', 'wonderlic', '40_yard', 'bench_press', 'vert_leap', 'broad_jump', 'shuttle', '3_cone']
    final_result = []
    for row in table_rows:
        result = dict()
        data = row.select('td')
        for i in range(len(data)): 
            result[types_data[i]] = data[i].get_text()
        final_result.append(result)
        print(result)
    return final_result

if __name__ == '__main__':
    main()
```

```python
df = pd.read_csv('combine_data.csv')
weight_percentiles = []
df = df[df['weight'].notna()]
for idx, player in df.iterrows():
    weight = player['weight']
    percentile = len(df[weight > df['weight']]) / (len(df) - 1)
    weight_percentiles.append(percentile)
df['weight_percentile'] = weight_percentiles
df.to_csv('combine_data.csv')
df.nlargest(10, 'weight_percentile')
```

```python
import pandas as pd
import math
df = pd.read_csv('combine_data.csv')
percentiles = dict()
wonderlic = df['wonderlic']
df.drop('wonderlic', axis=1)
for col in ['40_yard', 'bench_press', 'vert_leap', 'broad_jump', 'shuttle', '3_cone']:
    percentiles[col + '_percentile'] = []
for idx, player in df.iterrows():
    pos_df = df[(df['pos'] == player['pos']) & (df['year'] <= player['year'])]
    for col in ['40_yard', 'bench_press', 'vert_leap', 'broad_jump', 'shuttle', '3_cone']:
        value = player[col]
        col_pos_df = pos_df[pos_df[col].notna()]
        if math.isnan(value):
            percentile = value
        elif col == '40_yard' or col == 'shuttle' or col == '3_cone':
            percentile = len(col_pos_df[value < col_pos_df[col]]) / max(1, (len(col_pos_df) - 1))
        else:
            percentile = len(col_pos_df[value > col_pos_df[col]]) / max(1, (len(col_pos_df) - 1))
        percentiles[col + '_percentile'].append(percentile)

for col in ['40_yard', 'bench_press', 'vert_leap', 'broad_jump', 'shuttle', '3_cone']:
    df[col + '_percentile'] = percentiles[col + '_percentile']
display(df)
df['wonderlic'] = wonderlic
df.to_csv('combine_data.csv')
print('done')
```
